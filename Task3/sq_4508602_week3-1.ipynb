```python
#question 3.2
from math import *
from scipy.constants import G

#Lists with masses, radii and names
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

#Create the Dictionary
escapeVelocities = {}

#Calculate the escape velocity and add them to the dictionary
for i in range(len(masses)):
    print(f"{masses[i]:.2e}")
    escVel = sqrt((2*G*masses[i])/radii[i])
    escapeVelocities.update({names[i] : escVel})
print(escapeVelocities)
```

```python
#question 3.17
import numpy as np
from math import *

#insert your value M
Number = int(input("Input a value:"))
randomArray = np.random.uniform(0,2, Number)

#function
def stdev (N, randomArray):
    M_sum = 0
    std = 0
    
    #calculate the mean
    for i in range(0, N):
        M_sum += randomArray[i]
        mean = M_sum/N
        
    #calculcate the standard deviation
    for i in range(0, N):
        std += (randomArray[i]-mean)**2
        sigma = (std/N)**0.5
            
    return sigma

std_d = ((1/Number)*(np.sum((randomArray)-(1/Number)*np.sum(randomArray))**2))**0.5

print(stdev(Number, randomArray))
print(np.std(randomArray))
print(std_d)
```
